# Exercises with NetworkX

1. Graph with x nodes at random position
2. Graph with x nodes at random position and all nodes are connected with edges
3. Graph with x nodes at random position, nodes with given distance are connected


Following the results from the lecture discussion


```python
import math
import networkx as nx
import random
import matplotlib.pylab as plt
%matplotlib inline
```

```python
# Anzahl der Knoten
num_nodes = 50
distance_threshold = 30  # Threshold für Verbindung (z.B. 10 Einheiten)

# Zufällige Positionen für Knoten erzeugen
positions = {i: [random.uniform(0, 80), random.uniform(0, 80)] for i in range(num_nodes)}
```

```python
# -------- Vollständig verbundener Graph --------
G_full = nx.Graph()
for i in range(num_nodes):
    G_full.add_node(i, mypos=positions[i])
for i in range(num_nodes):
    for j in range(i + 1, num_nodes):
        G_full.add_edge(i, j)

# Plot für vollständig verbundenen Graph
plt.figure(figsize=(8, 8))
nx.draw(G_full, pos=positions, with_labels=True, node_size=100)
plt.title("Vollständig verbundener Graph")
plt.show()
```

```python
# -------- Graph mit Distanz-Threshold --------
G_thresh = nx.Graph()
for i in range(num_nodes):
    G_thresh.add_node(i, mypos=positions[i])
for i in range(num_nodes):
    for j in range(i + 1, num_nodes):
        pos_i = positions[i]
        pos_j = positions[j]
        distance = math.dist(pos_i, pos_j)
        if distance < distance_threshold:
            G_thresh.add_edge(i, j)

# Plot für Distanz-basierter Graph
plt.figure(figsize=(8, 8))
nx.draw(G_thresh, pos=positions, with_labels=True, node_size=100)
plt.title(f"Graph mit Distanz-Threshold = {distance_threshold}")
plt.show()
```
